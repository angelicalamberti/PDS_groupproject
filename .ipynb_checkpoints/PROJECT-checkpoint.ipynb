```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#1: DATA WRANGLING AND ANALYSIS

data= pd.read_csv('city_data.csv', sep='|', header=1, parse_dates=['Last Data Update'])
#I want the dates to be in a datetime64 format 

print(data.shape)
data.head()
#the dataset has 86 rows and 14 columns
```

```python
#We have 80 missing values for Average Price Groceries, 1 for Population Density, Working Age Population, Unemployment Rate, GDP per Capita, Main Spoken Languages
```

```python
#I want to delete duplicate info, but before I check info are the same
data.set_index('City').loc['Berlin. Germany']
data.set_index('City').loc['Malaga, Spain']
```

```python
#missing values treatment: drop them 
missing_values = data.isnull().sum()
data = data.dropna(axis=1, how='all')   
```

```python
#duplicates treatment: drop duplicates
data.drop_duplicates(inplace=True)
data
```

```python
data['City'].describe()
#no more duplicates info
```

```python
#2. a) Which country appears most frequently in the dataset? GERMANY
data_copy=data #We can work on the copy of data 
data_copy[['city', 'country']] = data_copy['City'].str.split(',', expand=True)
countries_count = data_copy['country'].value_counts()
countries_count

#How many cities are associated with it? 9 cities are associated with Germany 
```

```python
#b) How many cities are present in total? How many are associated with Greece? 
#problem with Greece: 'Athens' is in the country column, I have to change it
#I need to expand some rows where the separator in the city column was not a ','
```

```python
data_copy.info ()
```

```python
data_copy['city'].tolist()
```

```python
data_copy['country'].tolist()
#I have 2 missing values
```

```python
data_copy[data_copy['country'].isnull()]
#this is because in line 12 we have a ';' as a separator, and in the line 15 we have a '.'
```

```python
line_12=data_copy.iloc[12]
line_15=data_copy.iloc[15]
data_copy[['city', 'country']].iloc[12] = line_12.str.split(';', expand=True)
data_copy[['city', 'country']].iloc[15] = line_15.str.split('.', expand=True)
data_copy = data_copy.rename(columns={'City': 'city&country'})
data_copy
```

```python
#how many cities are present in total? 84 cities
cities_sum = data_copy['city'].value_counts().sum()
print(cities_sum)
#how many are associated with Greece? 2: Athens and Zagreb
cities_count=data_copy['country'].value_counts()
print(cities_count)
```

```python
#3. a) Entries uploaded before April 2023 need to be updated. Which cities would require an update? 
begin_April_23=pd.to_datetime('2023-04-1')
filter=data_copy[data_copy['Last Data Update']< begin_April_23]
filter['city']
#Basel, Munich, Frankfurt am Main, Lyon, Venice and Gothenburg need an update
```

```python
#b) How many days ago was the last update? On what day, month, and year did it occur? 
date_today= pd.Timestamp('today').normalize()
last_update=max(data_copy['Last Data Update'])
days= date_today-last_update
print(days)
#The last update was 341 days ago 
print(last_update)
#The last update was on 2024-12-12
```
